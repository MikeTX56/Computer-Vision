#### 2. Displaying the color image using matplotlib

```python
plt.imshow(avacado_img)
plt.axis("off")
plt.show()
```

```python
### Because its BGR lets reverse the color image

avacado_img_channels_reversed = avacado_img[:,:,::-1]
plt.imshow(avacado_img_channels_reversed)
plt.axis("off")
plt.show()
```

```python
# spliting and merging color channels
# since its CV it must B, G, R
b, g,r = cv2.split(avacado_img)

plt.figure(figsize=[20, 5])

#red channel1
plt.subplot(1, 4, 1)
plt.imshow(r, cmap = "gray")
plt.title("Red channel 1")
plt.axis("off")

#green channel1

plt.subplot(1, 4, 2)
plt.imshow(g, cmap = "gray")
plt.title("Green channel 1")
plt.axis("off")

#Blue channel 1
plt.subplot(1, 4, 3)
plt.imshow(b, cmap = "gray")
plt.title("Blue channel 1")

#merge he individualschannels into a BGR image

merged_img = cv2.merge((b, g, r))
plt.subplot(1, 4, 4)
plt.imshow(merged_img[:,:,::-1])
plt.title("Merge image output")
plt.axis("off")
```
