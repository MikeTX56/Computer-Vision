```python
import os
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline
```

#### Displaying the color image using matplotlib

```python
plt.imshow(avacado_img)
plt.axis("off")
plt.show()
```

```python
### Because its BGR lets reverse the color image

avacado_img_channels_reversed = avacado_img[:,:,::-1]
plt.imshow(avacado_img_channels_reversed)
plt.axis("off")
plt.show()
```

```python
# spliting and merging color channels
# since its CV it must B, G, R
b, g,r = cv2.split(avacado_img)

plt.figure(figsize=[20, 5])

#red channel1
plt.subplot(1, 4, 1)
plt.imshow(r, cmap = "gray")
plt.title("Red channel 1")
plt.axis("off")

#green channel1

plt.subplot(1, 4, 2)
plt.imshow(g, cmap = "gray")
plt.title("Green channel 1")
plt.axis("off")

#Blue channel 1
plt.subplot(1, 4, 3)
plt.imshow(b, cmap = "gray")
plt.title("Blue channel 1")

#merge he individualschannels into a BGR image

merged_img = cv2.merge((b, g, r))
plt.subplot(1, 4, 4)
plt.imshow(merged_img[:,:,::-1])
plt.title("Merge image output")
plt.axis("off")
```

#### changing from BGR to RGB color space

```python
avacado_img_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)

#converting BGR to RGB

avacado_img_rgb = cv2.cvtColor(avacado_img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(avacado_img_rgb)
plt.title("BGR to RGB")
plt.axis("off")
plt.show()
```

#### changing BGR to HSV (hue , saturation, value)

```python
avacado_img_hsv = cv2.cvtColor(avacado_img_bgr, cv2.COLOR_BGR2HSV)

# split the immage into H, S, V components

h, s, v =cv2.split(avacado_img_hsv)

#show the channel
plt.figure(figsize=[20, 5])

#Hue channel 1
plt.subplot(1, 4, 1)
plt.imshow(h, cmap = "gray")
plt.title("H channel")
plt.axis("off")

#saturation channel 1
plt.subplot(1,4,2)
plt.imshow(s, cmap = "gray")
plt.title("saturation channel")
plt.axis("off")

#value channel
plt.subplot(1, 4, 3)
plt.imshow(v,cmap = "gray")
plt.title("value channel")
plt.axis("off")

#original image
plt.subplot(1, 4, 4)
plt.imshow(avacado_img_rgb, cmap = "gray")
plt.title("Original image")
plt.axis("off")
```

```python
#converting BGR o Grayscale
avacado_img_grayscale = cv2.cvtColor(avacado_img_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(avacado_img_grayscale, cmap = "gray")
plt.title("BGR to Grayscale")
plt.axis("off")
plt.show()
```
